# Plotting in Julia

- ## Time Series

```julia
using TimeSeries,Plots,Dates,Random
pyplot()
dates = Date(1980, 1, 1):Day(1):Date(1980, 12, 31)
ta = TimeArray(dates, rand(20:31,length(dates)))
plot(ta)
```

- ## Heat Map

```julia
using RDatasets,Plots,DataFrames
pyplot()
df = dataset("datasets","Titanic")
class = unique(df.Class)
age = unique(df.Age)
z = float((1:4) * reshape(1:length(age), 1, :))
heatmap( age,class, z, aspect_ratio=1)
```

- ## Scatter Plot

```julia
using RDatasets,Plots
pyplot()
df = dataset("datasets","iris")
scatter(df.PetalLength,df.PetalWidth,df.Species,markersize=20)
```

```julia
using StatsPlots
using RDatasets
iris = dataset("datasets", "iris")
@df iris scatter(:PetalWidth, :PetalLength, group = :Species,markersize=15)
```

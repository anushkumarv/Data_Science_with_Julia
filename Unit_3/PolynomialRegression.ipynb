```julia
# load dataset. Here we will use built-in Boston housing prices dataset using RDatasets package
boston_housing_prices = dataset("MASS","Boston")
```

```julia
x = boston_housing_prices[:,13] # using col LStat as x
y = boston_housing_prices[:,14] # using col medv as y
```

Polynomial regression is just like linear regression except here we try to train for the equation <br>
f(x) = w1*x^n + w2*x^(n-1) + w3*x^(n-2) + ... + wn 

```julia
# Lets consider a simple case where we try to fit f(x) = a*x^2 + b
w1 = [rand()]
b = [rand()]
f_2nd_deg_simple(x) = w1 .* x .* x .+ b
function squared_loss_2nd_deg_simple(x)
    y_hat = f_2nd_deg_simple(x)
    return sum((y .- y_hat).^2) / 2
end
function train_simple_2nd_deg_simple(x)
    grads = gradient(()->squared_loss_2nd_deg_simple(x), params(w1,b))
    w1 .-= 0.00000001 * grads[w1]
    b .-= 0.00000001 * grads[b]
end
```

```julia
println("Before training")
println(w1," ",b)
println(squared_loss_2nd_deg_simple(x))
```

```julia
for i in 1:800000
    train_simple_2nd_deg_simple(x)
end
```

```julia
println("After training")
println(w1," ",b)
println(squared_loss_2nd_deg_simple(x))
```

```julia
x_generated = [item for item in 0:0.1:40]
y_predicted = f_2nd_deg_simple(x_generated)
println(length(y_predicted), length(x_generated))
# len(x_generated)
scatter(x,y)
plot!(x_generated,y_predicted,linewidth=3)
```

```julia
using Flux: @epochs
```

```julia
# We can similarly train higher and higher polynomials
# Lets have a look at that
# Initialzie with random weights
w1 = [rand()]
w2 = [rand()]
w3 = [rand()]
w4 = [rand()]
b = [rand()]
```

```julia
# train functions
f_2nd_deg(x) = w1.*x.^2 .+ w2.*x  .+ b
f_3rd_deg(x) = w1.*x.^3 .+ w2.*x.^2 .+ w3.*x .+ b
f_4th_deg(x) = w1.*x.^4 .+ w2.*x.^3 .+ w3.*x.^2 .+ w4.*x .+ b
```

```julia
#loss function
function squared_loss(x,y,f)
    y_hat = f(x)
    return sum((y .- y_hat).^2) / 2
end
```

```julia
#train function
function train_custom(x, y, f, ps, eta)
    grads = gradient(()->squared_loss(x, y, f), ps)
    for x in ps
        Flux.Optimise.update!(x, -grads[x] * eta)
    end 
end
```

```julia
# lets write a function to reset weights
function reset_weights()
    w1 = [randn()]
    w2 = [randn()]
    w3 = [randn()]
    w4 = [randn()]
    b = [randn()]
end
```

```julia
# lets train 3rd deg polynomial
parameters_3rd_deg = params(w1,w2,b)
learning_rate = 0.0000000001
println("Weghts before training")
println(w1," ",w2," ",w3," ",b)
println("Loss before training")
println(squared_loss(x,y,f_3rd_deg))
@epochs 50 train_custom(x,y,f_2nd_deg,parameters_3rd_deg,learning_rate);
println("Loss after training")
println(squared_loss(x,y,f_3rd_deg))
```
